```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from your dataframe
data = pd.read_csv("some_data.csv")
```

```python
# Select the columns to plot
x = data['_BMI5']
y = data['GENHLTH']
```

```python
# Create a scatter plot
plt.scatter(x, y, alpha=0.5)
```

```python
# Set the title and axis labels
plt.title('BMI vs General Health')
plt.xlabel('BMI')
plt.ylabel('General Health')
```

```python
# Show the plot
plt.show()
```
